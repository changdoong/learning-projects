```python
#데이터 불러오기
import pandas as pd

df = pd.read_csv('916 국가보훈대상자로서의 자긍심 정도.csv')
df
```

```python
# 고유값 확인
df['C1_NM'].unique()
```

```python
# C1_NM이 전체인 데이터 추출
df1 = df[df['C1_NM'] == '전체']
df1.head()
```

```python
#연도별 설문조사 파이차트 시각화
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'Malgun Gothic'

target_years = [2018, 2021, 2024]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, year in zip(axes, target_years):
    sub_df = df1[df1['연도'] == year]
    
    if not sub_df.empty:
        wedges, texts, autotexts = ax.pie(
            sub_df['DT'],
            labels=sub_df['C2_NM'],
            autopct='%.1f%%',
            textprops={'fontsize': 14},    
        )
        ax.set_title(f'{year}년 분포', fontsize=20)

plt.tight_layout()
plt.show()
```
