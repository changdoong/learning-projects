```python
#라이브러리, csv 파일 불러오기
import pandas as pd
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("D:/learning-projects/soldier_df/news/네이버뉴스/이재명 정부_국가유공자_네이버뉴스 크롤링.csv")
df
```

```python
#불용어 처리 및 워드 클라우드 시각화
from konlpy.tag import Okt
from collections import Counter


wc_dict = df.set_index('제목').to_dict()['본문']
text = " ".join(wc_dict.values()).replace('\n', ' ').replace('\r', ' ')


okt = Okt()
tokens = okt.morphs(text)

stop_words = ['을', '를', '이', '가', '은', '는', '에', '의', '와', '과', 
              '에서', '으로', '도', '로', '하고', '보다', '등', '대한', 
              '있다', '없는', '위해', '하여', '하게', '하며', '했다', '된다',
              '밝혔다','더욱','달리','함께','하는','까지','않은','에게',
              '따르면','됐다','하기','현재','있도록','받고','이다','부터',
              '이번','됩니다','뉴시스','우선','통한','있는','하겠다','통해',
              '가능한','해당','최초','또는','맞아','분들','찾아','이후','품은','했다고']
tokens_cleaned = [t for t in tokens if t not in stop_words and len(t) > 1]

word_freq = Counter(tokens_cleaned)
top_50_words = dict(word_freq.most_common(50))


mask_path = "D:/learning-projects/soldier_df/news/네이버뉴스/문재인2.png"
mask_image = np.array(Image.open(mask_path).convert("L"))


threshold = 180  
binary_mask = np.where(mask_image < threshold, 0, 255).astype(np.uint8)


font_path = 'C:/Windows/Fonts/malgun.ttf'  
wc_img = WordCloud(
    font_path=font_path,
    background_color='white',
    width=800,
    height=400,
    mask=binary_mask,
    contour_width=1,
    contour_color='black'
).generate_from_frequencies(top_50_words)


plt.figure(figsize=(12, 6))
plt.imshow(wc_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```
