```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

#  
df_income = pd.read_csv('   ().csv')
df_living = pd.read_csv('   .csv')

# 1.       ( )
# 2018: 297, 2021: 333, 2024: 521.5 (:  )
#        
national_income_data = []
for year in [2018, 2021, 2024]:
    if year == 2018:
        income = 297
    elif year == 2021:
        income = 333
    else:
        income = 521.5
    
    #  ''     
    for group in df_income[''].unique():
        national_income_data.append({
            '': group,
            ' ( )': income,
            '': year,
            ' ': '  '
        })

df_korean = pd.DataFrame(national_income_data)


# 2.    
df_income_filtered = df_income[['', ' ', '']].rename(columns={' ': ' ( )'})
df_income_filtered[' '] = ' '

# 3.     
df_living_filtered = df_living[df_living[''] == ' '].copy()
df_living_filtered = df_living_filtered.rename(columns={'': ' ( )'})
df_living_filtered[' '] = ' '
df_living_filtered = df_living_filtered[['', ' ( )', '', ' ']]


# 4.   
combined_df = pd.concat([df_income_filtered, df_living_filtered, df_korean])
combined_df[' ( )'] = combined_df[' ( )'].astype(float)


# 5.   
plt.figure(figsize=(15, 8))
sns.barplot(
    data=combined_df,
    x='',
    y=' ( )',
    hue=' ',
    palette='viridis',
    errorbar=None
)

#      
plt.title('        ', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel(' ( )', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.legend(title=' ', loc='upper left')
plt.ylim(0, 800)

#  
plt.tight_layout()
plt.show()
```
