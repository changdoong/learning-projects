```python
import pandas as pd

df = pd.read_csv('보훈대상자 평균 개인소득 총집합.csv')
df
```

```python
# 컬럼 명 변경
df = df.rename(columns={'소득': '연간 소득'})
```

```python
# 경상소득 데이터만 추출
dfs = df[df['소득종류'] == '전체(경상소득)']
dfs.head()
```

```python
# 경상소득 데이터 복사
dfs = df[df['소득종류'] == '전체(경상소득)'].copy()
dfs
```

```python
# 월간 소득 컬럼 생성 후 반올림
dfs['월간 소득'] = round(dfs['연간 소득'] / 12.0,1)
dfs
```

```python
csv 파일로 저장
dfs.to_csv('보훈대상자 평균 개인소득 총집합(월간 소득 컬럼 추가).csv',index=False)
```
