```python
#데이터 불러오기
import pandas as pd
df = pd.read_csv('경제활동 참여 상태 2021.csv')
df
```

```python
# 불필요한 컬럼 제거
df = df.drop(columns=['Unnamed: 0','C1_OBJ_NM','UNIT_NM','ITM_NM','C2_OBJ_NM','TBL_NM'])
df['연도'] = '2021'
df
```

```python
# 컬럼 이름 변경
df = df.rename(columns={
    "C1_NM": "구분",
    "C2_NM": "취업여부",
    "DT": "비율",
    "연도": "연도"
})
df
```

```python
# csv 파일로 저장
df.to_csv('경제활동 참여 상태 2021.csv', index=False)
```
