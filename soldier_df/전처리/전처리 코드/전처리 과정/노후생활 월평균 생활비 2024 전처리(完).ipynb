```python
df = pd.read_csv('노후생활 월평균 생활비 2024.csv')
df.head()
```

```python
# 불필요한 컬럼 제거
rem = ['Unnamed: 0','C1_OBJ_NM','UNIT_NM']

df.drop(columns = rem, inplace = True)
df
```

```python
# 인덱스 C1_NM으로 설정
df.set_index('C1_NM',inplace=True)
```

```python
df
```

```python
#인덱스 고유값 확인 
df.index.unique()
```

```python
# 특정 인덱스 애들 꺼내서 데이터프레임을 만듦
man = ['국가유공자(본인)', '국가유공자(유족)', '고엽제', '참전유공자', '특수임무유공자']

dfs = df.loc[df.index.isin(man)]
```

```python
dfs.index.unique()
```

```python
# 연도 컬럼 생성하기. 연도 옆에 row 수 만큼 곱해주면됌
연도 = [2024] * 15
```

```python
dfs['연도'] = 연도
```

```python
dfs
```

```python
dfs = dfs.rename(columns={
    "C2_NM": "소득종류",
    "DT": "비용",
    "ITM_NM": "생활비",
    "TBL_NM": "테이블명",
    "연도": "연도"
})
dfs.index.name = "구분"
dfs
```

```python
dfs.to_csv('노후생활 월평균 생활비 2024 전처리.csv')
```
