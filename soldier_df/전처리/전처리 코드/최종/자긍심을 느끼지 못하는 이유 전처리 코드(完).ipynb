```python
# 데이터 불러오기
import pandas as pd 
df = pd.read_csv('자긍심을 느끼지 않는 이유.csv')
df
```

```python
#2018년 기준 국가유공자 데이터 추출
df1 = df[df['대상'] == '국가유공자(본인)']
df1 = df1[df1['연도'] == 2018].drop(columns=['대상','비율'])
df1.set_index('연도',inplace=True)
```

```python
df1.to_csv('2018년 국가유공자 자긍심.csv')
df1
```

```python
# 2018년 국가유공자의 이유 컬럼만 남김
df2 = df[df['대상'] == '국가유공자(본인)']
df2 = df2[df2['연도'] == 2021].drop(columns=['대상','비율'])
df2.set_index('연도',inplace=True)
```

```python
# csv 파일 저장
df2.to_csv('2021년 국가유공자 자긍심.csv')
df2
```

```python
# 2024년 국가유공자의 이유 컬럼만 남김
df3 = df[df['대상'] == '국가유공자(본인)']
df3 = df3[df3['연도'] == 2024].drop(columns=['대상','비율'])
df3.set_index('연도',inplace=True)
```
