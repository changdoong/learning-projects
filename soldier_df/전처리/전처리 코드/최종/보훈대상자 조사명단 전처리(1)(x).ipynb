```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('보훈대상자 조사명단.csv')

# Create a mapping dictionary for '예' and '아니오'
# 1 = Yes, 2 = No
mapping = {1: '예', 2: '아니오'}

# Find all columns that end with '_인지여부'
recognition_columns = [col for col in df.columns if col.endswith('_인지여부')]

# Apply the mapping to each of the identified columns
for col in recognition_columns:
    df[col] = df[col].replace(mapping)

# Display a few of the changed columns to verify the result
print(df[recognition_columns].head())
```

```python
df.to_csv('보훈대상자 조사명단.csv')
```
