```python
# CSV 파일 불러오기
df = pd.read_csv('보훈대상자 조사명단.csv')

# 연령대 매핑 딕셔너리 정의
age_mapping = {
    5: '80대 이상',
    4: '70대',
    3: '60대',
    2: '50대',
    1: '40대 이하'
}

# '보훈대상자 연령대(5개 범주)' 컬럼의 값들을 딕셔너리에 따라 변경
df['보훈대상자 연령대(5개 범주)'] = df['보훈대상자 연령대(5개 범주)'].replace(age_mapping)

# 변경된 데이터 확인 (상위 5개 행 출력)
print(df[['보훈대상자 만나이', '보훈대상자 연령대(5개 범주)']].head())
df.to_csv('보훈대상자 조사명단.csv')
```

```python
# CSV 파일 불러오기
df = pd.read_csv('보훈대상자 조사명단.csv')

# '경제활동참여상태' 컬럼의 숫자(키)를 텍스트(값)로 매핑합니다.
status_mapping = {
    3: '일하지 않고 있으며 일할 의사도 없다',
    2: '일을 하지 않지만 일하고 싶다',
    1: '일을 한다'
}

# replace() 함수를 사용하여 컬럼 값 변경
df['경제활동참여상태'] = df['경제활동참여상태'].replace(status_mapping)

# 변경된 데이터 확인 
print(df[['경제활동참여상태']].head())
df.to_csv('보훈대상자 조사명단.csv')
```

```python
import pandas as pd

# 파일 불러오기
df = pd.read_csv('보훈대상자 조사명단.csv')

# 매핑할 딕셔너리 
mapping = {1: '예', 2: '아니오'}

# 데이터프레임 컬럼 중 '_인지여부'로 끝나는 컬럼만 추출
recognition_columns = [col for col in df.columns if col.endswith('_인지여부')]

# 매핑
for col in recognition_columns:
    df[col] = df[col].replace(mapping)

# 매핑 됐는지 확인
print(df[recognition_columns].head())
```

```python
df.to_csv('보훈대상자 조사명단.csv')
```
