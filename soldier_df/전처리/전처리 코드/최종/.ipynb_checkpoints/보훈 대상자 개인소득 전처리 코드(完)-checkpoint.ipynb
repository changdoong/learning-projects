```python
# 데이터 불러오기
import pandas as pd

df = pd.read_csv('보훈대상자 평균 개인소득_전체_통합.csv')
df.head()
```

```python
df
```

```python
#월 별 소득 컬럼 추가
df['월 별 소득'] = (df['DT'] / 12)
```

```python
#월 별 소득 컬럼 데이터 타입 변환
df['월 별 소득'] = df['월 별 소득'].astype(int)
```

```python
#DT 데이터타입 변환
df['DT'] = df['DT'].astype(int)
```

```python
#컬럼 명 순서 변경
df1 = df[['C1_NM','C2_NM','DT','월 별 소득','연도']]
```

```python
# 데이터 저장
df1.to_csv('보훈대상자 평균 개인소득_전체_통합 월별 소득 추가.csv')
```
