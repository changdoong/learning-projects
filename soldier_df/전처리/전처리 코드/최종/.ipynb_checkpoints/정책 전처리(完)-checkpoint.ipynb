```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv('정책_변환_결과.csv')
df2 = pd.read_csv('2020_시행계획_요약.csv')
df3 = pd.read_csv('2021_시행계획_요약.csv')
df4 = pd.read_csv('2025_핵심추진과제.csv')
```

```python
df2 = df2.rename(columns = {'제목':'정책'})
df3 = df3.rename(columns = {'제목':'정책'})
```

```python
df = pd.concat([df1, df2, df3, df4])
df = df.sort_values(by='연도')
df.to_csv('정책 통합.csv')
```
