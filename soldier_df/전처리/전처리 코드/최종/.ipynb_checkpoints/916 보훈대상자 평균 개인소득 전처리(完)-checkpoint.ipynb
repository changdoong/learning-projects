```python
import pandas as pd

df1 = pd.read_csv('보훈대상자 평균 개인소득 2018.csv')
df2 = pd.read_csv('보훈대상자 평균 개인소득 2021.csv')
df3 = pd.read_csv('보훈대상자 평균 개인소득 2024.csv')
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df1.set_index('C1_NM',inplace=True)
```

```python
df1.drop(columns = ['Unnamed: 0','C1_OBJ_NM','UNIT_NM','ITM_NM','C2_OBJ_NM','TBL_NM'],inplace=True)
```

```python
df1['연도'] = 2018
```

```python
df1
```

```python
df2.set_index('C1_NM',inplace=True)
```

```python
df2.drop(columns = ['Unnamed: 0','C1_OBJ_NM','UNIT_NM','ITM_NM','C2_OBJ_NM','TBL_NM'],inplace=True)
```

```python
df2['연도'] = 2021
```

```python
df2
```

```python
df3.set_index('C1_NM',inplace=True)
```

```python
df3.drop(columns = ['Unnamed: 0','C1_OBJ_NM','UNIT_NM','ITM_NM','C2_OBJ_NM','TBL_NM'],inplace=True)
```

```python
df3['연도'] = 2024
```

```python
df3
```

```python
dfa = pd.concat([df1,df2,df3])
```

```python
dfa.to_csv('916 보훈대상자 평균 개인소득.csv')
```
