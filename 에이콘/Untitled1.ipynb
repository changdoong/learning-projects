```python
import numpy as np
import math
from scipy import stats
```

```python
a_scores = [85,90,88,92,75,79,95,89,84,91,97,90,86,93,88]
b_scores = [78,82,80,76,85,79,81,84,77,8,80,82,78,79,81,85,80]

t_stat, p_value = stats.ttest_ind(a_scores,b_scores, equal_var=False)
print(f't 통계랑 : {t_stat:.3f}')
print(f'p 값 : {p_value:.4f}')

alpha = 0.05
if p_value < alpha:
    print('두 반의 평균 점수 차이는 통계적으로 유의미')
else:
    print('두 반의 평균 점수 차이는 통계적으로 유의미하지 않다')
    
```
