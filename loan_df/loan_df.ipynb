```python
ohe_df = pd.get_dummies(train_df, columns=['loan_intent'])
```

```python
ohe_df['loan_grade'].unique()
```

```python
ohe_df['loan_grade'] = ohe_df['loan_grade'].map({
    'A' : 7,
    'B' : 6,
    'C' : 5,
    'D' : 4,
    'E' : 3,
    'F' : 2,
    'G' : 1
})
```

```python
ohe_df2 = pd.get_dummies(ohe_df,columns = ['person_home_ownership'])
```

```python
ohe_df2.dtypes
```

```python
ohe_df2['cb_person_default_on_file'] = ohe_df2['cb_person_default_on_file'].map({
    'N' : 0,
    'Y' : 1
})
```

```python
X = ohe_df2.drop(columns=['id','loan_status'])
y = ohe_df2['loan_status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=2020158011)
```
