```python
genus = {'Crocodylus' : 0,
         'Osteolaemus' : 1,
         'Mecistops' : 2
        }
df['Genus'] = df['Genus'].map(genus)
```

```python
df['Genus'].unique()
```

```python
from category_encoders import BinaryEncoder

binary_encoder = BinaryEncoder(cols='Country/Region')
df_encoded = binary_encoder.fit_transform(df)
df_encoded
```

```python
df['Habitat Type'].unique()
```

```python
binary_encoder2 = BinaryEncoder(cols = 'Habitat Type')
df_encoded1 = binary_encoder2.fit_transform(df_encoded)
df_encoded1
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=2020158011)
kmeans.fit(df_encoded1)
```

```python
print(kmeans.labels_)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_encoded1)

labels = kmeans.labels_

plt.figure(figsize=(8,6))
scatter = plt.scatter(df_pca[:,0], df_pca[:,1], c=labels, cmap='viridis', alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('KMeans Clustering PCA')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```
