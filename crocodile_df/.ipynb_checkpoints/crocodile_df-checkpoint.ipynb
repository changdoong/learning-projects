```python
encoder = ce.BinaryEncoder(cols=['Country/Region'])
df_encoded = encoder.fit_transform(df)
df_encoded
```

```python
encoder1 = ce.BinaryEncoder(cols=['Habitat Type'])
df_encoded1 = encoder1.fit_transform(df_encoded)
df_encoded1
```

```python
df_encoded1['Genus'].unique()
```

```python
map_list = {'Crocodylus' : 0,
            'Osteolaemus' : 1,
            'Mecistops' : 2}
```

```python
df_encoded1['Genus'] = df_encoded1['Genus'].map(map_list)
```

```python
df_encoded1
```

```python
X = df_encoded1.drop(columns='Genus')
y = df_encoded1['Genus']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.2)
```

```python
from lazypredict.Supervised import LazyClassifier

clf = LazyClassifier(verbose=0,ignore_warnings=True)
models, predictions = clf.fit(X_train, X_val, y_train, y_val)

print(models)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
import optuna
```

```python
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 50, 500),
        "max_depth": trial.suggest_int("max_depth", 3, 50),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 20),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 20),
        "max_features": trial.suggest_categorical("max_features", ["sqrt", "log2", None]),
        "criterion": trial.suggest_categorical("criterion", ["gini", "entropy", "log_loss"])
    }

    model = ExtraTreesClassifier(**params, n_jobs=-1, random_state=2020158011)

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=2020158011)
    score = cross_val_score(model, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1).mean()

    return -score
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50)

print("Best trial:", study.best_trial.params)
print("Best score:", -study.best_trial.value)  # 음수로 넣었으니까 다시 뒤집어줌

# study = optuna.create_study(direction='minimize')
# study.optimize(objective, n_trials=50)
# print('Best trial:', study.best_params)
```

```python
best_params = study.best_trial.params
best_model = ExtraTreesClassifier(**best_params, n_jobs=-1, random_state=2020158011)
best_model.fit(X_train, y_train)
```
