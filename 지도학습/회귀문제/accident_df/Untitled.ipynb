```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler

scaled = StandardScaler().fit_transform(train_df.iloc[:,:-1])
```

```python
scaled.mean(axis=0)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
pca.fit(scaled)
accident_pca = pca.transform(scaled)
accident_pca.shape
```

```python
pca_df = pd.DataFrame(accident_pca, columns=['pca_comp1','pca_comp2','pca_comp3','pca_comp4','pca_comp5','pca_comp6','pca_comp7','pca_comp8','pca_comp9','pca_comp10'])
pca_df['accident_risk'] = train_df['accident_risk']

pca_df.head(10)
```

```python
pca.explained_variance_ratio_
```

```python
X = pca_df.drop(columns='accident_risk')
y = pca_df['accident_risk']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
rf.predict(X_val)
mse = mean_squared_error(y_val,rf.predict(X_val))
mae = mean_absolute_error(y_val, rf.predict(X_val))
r2_score = r2_score(y_val, rf.predict(X_val))
print(f' mse : {mse:.4f}, mae : {mae:.4f}, r2_score : {r2_score}')
```
