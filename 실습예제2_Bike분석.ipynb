#### features

- datetime - hourly date + timestamp  
- season -  1 = spring, 2 = summer, 3 = fall, 4 = winter
- holiday - whether the day is considered a holiday
- workingday - whether the day is neither a weekend nor holiday
- weather
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp - temperature in Celsius
- atemp - "feels like" temperature in Celsius
- humidity - relative humidity
- windspeed - wind speed
- casual - number of non-registered user rentals initiated
- registered - number of registered user rentals initiated
- count - number of total rentals

## 4. 통계적 가설검정

### 날씨(weather)에 따라 자전거 대여수(count)는 차이가 있는가?

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = smf.ols('count ~ C(weather)',data=df).fit()
weather_bike = sm.stats.anova_lm(model, typ=2)
print(weather_bike)
```

- 날씨에 따른 변동이 큼
- 날씨와 자전거 대여 수는 서로 독립이다라는 귀무가설 기각 

### 근무여부(workingday)에 따라 자전거 대여수(count)는 차이가 있는가?

```python
model = smf.ols('count ~ C(workingday)', data=df).fit()
workingday_count = sm.stats.anova_lm(model, typ=2)
print(workingday_count)
```

- 근무일 여부는 자전거 대여수에 영향이 그렇게 없음
- 서로 독립이다 라는 귀무가설 채택

### 계절(season)에 따라 자전거대여수(count)는 차이가 있는가?

```python
model = smf.ols('count ~ C(season)', data=df).fit()
season_count = sm.stats.anova_lm(model, typ=2)
print(season_count)
```

- 계절은 자전거 대여수에 많은 영향을 끼침
- 두 변수는 독립이다 라는 귀무가설 기각
