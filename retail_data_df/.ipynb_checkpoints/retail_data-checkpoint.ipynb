### 결제 ID와 상품 정보만 확인

```python
retail_data_df2 = retail_data_df[['OrderID','ProdName']]
retail_data_df2.head()
```

#### groupby()를 사용해 결제ID 별로 구매상품 목록을 묶음
#### apply(list)를 사용해 기준이 되는 컬럼(ProdName)을 리스트 형태로 묶음

```python
basket_df = retail_data_df2.groupby('OrderID')['ProdName'].apply(list).reset_index()
basket_df
```

```python
df = pd.read_csv('retail_data.csv')

basket_df = retail_data_df.groupby('OrderID')['ProdName'].apply(list).reset_index()
basket_df.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_result = te.fit_transform(basket_df['ProdName']) # transform(데이터를 원핫인코딩으로 바꿔줌)
```

```python
import numpy as np
te_df = pd.DataFrame(te_result, columns=te.columns_)
te_df.head()
```

```python
te_df.shape #270개의 결제 건에 대해 1343개의 상품 존재 여부
```

## Apriori 알고리즘 구현

```python
from mlxtend.frequent_patterns import apriori

apriori(te_df, min_support=0.05) #최소 지지도 0.05 설정, 3363개의 빈발 항목 집합 반환
```

```python
frequent_itemsets = apriori(te_df, min_support=0.06, use_colnames=True) #최소 지지도를 0.06으로 올림
frequent_itemsets #61개의 빈발 항목집합
```

## 연관 규칙 추출 및 평가

```python
from mlxtend.frequent_patterns import association_rules 

association_rules(frequent_itemsets, metric='confidence', min_threshold=0.8)#association_rules() 빈발 항목 집합에서 연관 규칙을 추출
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.8)
rules.sort_values(by='lift', ascending=False).head() #내림차순 정렬
```
